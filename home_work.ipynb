```python
import pandas as pd

# Укажите путь к вашему файлу
file_path = 'web_clients_correct.csv'

# Загрузка данных из CSV
data = pd.read_csv(file_path)

# Функция для генерации текстового описания
def generate_description(row):
    pronoun = "совершил" if row["sex"] == "male" else "совершила"
    description = (
        f"Пользователь {row['name']} {row['sex']}ого пола, {int(row['age'])} лет {pronoun} "
        f"покупку на {row['bill']} у.е. с {row['device_type']} браузера {row['browser']}. "
        f"Регион, из которого совершалась покупка: {row['region']}."
    )
    return description

# Применение функции ко всем строкам
def generate_descriptions(data):
    return [generate_description(row) for _, row in data.iterrows()]

# Генерация описаний
descriptions = generate_descriptions(data)

# Сохранение в текстовый файл
output_path = 'customer_descriptions.txt'
with open(output_path, 'w', encoding='utf-8') as f:
    f.write("\n\n".join(descriptions))

# Вывод результатов в консоль
print("Сгенерированные описания:\n")
for description in descriptions:
    print(description)
    print("-" * 80)  # Разделитель для удобства чтения

print(f"\nФайл сохранен: {output_path}")
```
